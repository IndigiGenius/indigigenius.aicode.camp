# How should we prepare our data to be used for training?

## Saving our Original Data

After organizing all of your image data into correctly named folders, zip up that folder and put that zip file aside in a new folder labeled "Data Recovery Place". Zipping a folder up keeps the old folders in their place, and creates a new file that contains all of the information.

To zip a folder
1. select the folder by clicking on it
1. right-click on the folder
1. Select `Compress to ZIP File`



We want to set our original data aside in case we need to get back to the original stuff. We will modify the folders that are left (unzipped) to be better prepared to transfer into the docker container for training, while always having a backup in the "Data Recovery Place" folder.

## Reducing the Size of the Data

We don't need high definition images for training a model. Most models require that the images be of a smaller size. To accomplish this task we will use the `PowerToys` application.

We installed `PowerToys` in the first lesson for this unit, so refer to that notebook if it is not on your computer.

Follow the instructions below to resize all of your images.
1. Open one of the labeled folders with the images you want to resize.
1. Change the view to "Details".
1. Select one file and press `CTRL+A`, this will select all of the files in that folder.
1. Right click on one of the highlighted files while keeping all images highlighted within the folder.
1. Select `Resize with Image Resizer`
1. A window will open for the `Image Resizer` application.
    1. Select *Phone*
    1. Un-select *Ignore the orientation*
    1. Select *Overwrite* **(Note: this will overwrite all of your pictures, don't take this step if you have not put your images in the "Data Recover Place")**
1. Then click `Resize` (you may need to refresh the window to see the reduced sizes)




## Creating a ZIP File of the Resized Data

Once all your images are resized and in labeled folders, go to the parent folder and add "_RESIZED" to the end of the folder name. Then create a new zip file of the folder with the resized data. This should be a collection of labeled folders containing images of that label, but significantly smaller in size that we can then use for training.
